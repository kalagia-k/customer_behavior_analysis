```python
df["Review Rating"] = (df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median())))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df = df.rename(columns = {"purchase_amount_(usd)": "purchase_amount"})
```

```python
df.columns
```

```python
#create a column age group
labels = ["Young Adult", "Adult", "Middle-aged", "Senior"]
df["age_group"] = pd.qcut(df["age"], q = 4, labels = labels)
```

```python
df[["age", "age_group"]]
```

```python
# create purchase_frequency_days

frequency_mapping = {
    "Fortnightly": 14,
    "Weekly": 7,
    "Monthly": 30,
    "Bi-weekly": 14,
    "Annually": 365,
    "Every 3 Months": 90
}

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days", "frequency_of_purchases"]]
```

```python
from sqlalchemy import create_engine, text
import urllib
```

```python
server = r"localhost\SQLEXPRESS"
database = "master"   # <-- IMPORTANT for the test
driver = "ODBC Driver 17 for SQL Server"

params = urllib.parse.quote_plus(
    f"DRIVER={{{driver}}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    "Trusted_Connection=yes;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

with engine.connect() as conn:
    print(conn.execute(text("SELECT @@SERVERNAME")).scalar())
    print(conn.execute(text("SELECT DB_NAME()")).scalar())
    print(conn.execute(text("SELECT name FROM sys.databases WHERE name='customer_behavior'")).fetchall())
    print(conn.execute(text("SELECT HAS_DBACCESS('customer_behavior')")).scalar())
```

```python
# Step 1: Connwct to SQL

server = "Konstantina"
database = "customer_behavior"
driver = "ODBC Driver 17 for SQL Server"

params = urllib.parse.quote_plus(
    f"DRIVER={{{driver}}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    "Trusted_Connection=yes;"
)

engine = create_engine(
    f"mssql+pyodbc:///?odbc_connect={params}"
)

# Step 2: Load DataFrame into SQL

table_name = "customer"

df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
